```python
%matplotlib widget

# Imports
import lumicks.pylake as lk

from os.path import join
from ctraptools.kymos.interactive import FrameRangeSelector
```

```python
# Load the kymograph
filepath = ""
filename = ""
h5_file = lk.File(join(filepath, filename))

kymo_name = list(h5_file.kymos)[0]
kymo = h5_file.kymos[kymo_name]
```

```python
# Displaying the frame range selector
frs = FrameRangeSelector(kymo,px_min=0,px_max=50)
```

```python
# Displaying the chosen frame range
(x_min, x_max) = frs.get_range()
print("Minimum: ",x_min)
print("Maximum: ",x_max)
```
